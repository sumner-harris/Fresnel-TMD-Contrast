```python
import numpy as np
import matplotlib.pyplot as plt

import Fresnel_TMD

# set constants 
wavelength = 632.992
incident_angle = 32.5

TMD_layer_thickness = 0.7
SiO2_thickness = 90

# Refractive index for each layer at the given wavelength
TMD_index = 4.42-0.603j #WSe2
SiO2_index = 1.4689
Si_index = 3.872-0.01637j

# Input constants and calculate the contrast
# N is layer number, Cp, Cs, and Cnp are the contrast for p, s, and randomly polarizared light.
N, Cp, Cs, Cnp,_ = Fresnel_TMD.Fresnel_TMD(laser_wavelength_nm = wavelength,
                          incident_angle_degrees = incident_angle,
                          TMD_layer_thickness_nm = TMD_layer_thickness,
                          SiO2_thickness_nm = SiO2_thickness,
                          TMD_index = TMD_index,
                          SiO2_index = SiO2_index,
                          Si_index = Si_index)

# Plot the data
fig,ax = plt.subplots(dpi=300)

ax.plot(N, Cnp,'o-')

# Plot limits
ax.set_xlim(0,10)
ax.set_ylim(-1,0.1)

# Plot labels
ax.set_xlabel('Number of Layers')
ax.set_ylabel('Contrast')
```
